```python
from IPython.display import display_html
def display_side_by_side(*args):
    html_str=''
    for df in args:
        html_str+=df.to_html()
    display_html(html_str.replace('table','table style="display:inline"'),raw=True)

class display(object):
    """Display HTML representation of multiple objects"""
    template = """<div style="float: left; padding: 10px;">
    <p style='font-family:"Courier New", Courier, monospace'>{0}</p>{1}
    </div>"""
    def __init__(self, *args):
        self.args = args
        
    def _repr_html_(self):
        return '\n'.join(self.template.format(a, eval(a)._repr_html_())
                         for a in self.args)
    
    def __repr__(self):
        return '\n\n'.join(a + '\n' + repr(eval(a))
                           for a in self.args)
    
```

#### MERGE Method
>One-to-one joins

```python
df_1 = pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'group': ['Accounting', 'Engineering', 'Engineering', 'HR']})

df_2 = pd.DataFrame({'employee': ['Lisa', 'Bob', 'Jake', 'Sue'],
                    'hire_date': [2004, 2008, 2012, 2014]})

df_4 = pd.DataFrame({'group': ['Accounting', 'Engineering', 'HR'],
                    'supervisor': ['Carly', 'Guido', 'Steve']})

df_5 = pd.DataFrame({'group': ['Accounting', 'Accounting',
                              'Engineering', 'Engineering', 'HR', 'HR'],
                    'skills': ['math', 'spreadsheets', 'coding', 'linux',
                               'spreadsheets', 'organization']})

df_6 = pd.DataFrame({'name': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'salary': [70000, 80000, 120000, 90000]})
```

```python
display('df_1', 'df_2')
```

```python
df_3=pd.merge(df_1,df_2)
df_3
```

#### Many-to-one joins

```python
display('df_3', 'df_4')
```

```python
pd.merge(df_3, df_4)
```

#### Many-to-many joins

```python
display('df_1', 'df_5')
```

```python
pd.merge(df_1, df_5)
```

#### Specification of the Merge Key

```python
display('df_1', 'df_2', 'df_4', 'df_5')
```

```python
pd.merge(df_1, df_2, on='employee')
```

#### The left_on and right_on keywords

```python
display('df_1', 'df_6')
```

```python
pd.merge(df_1, df_6, left_on='employee', right_on='name')
```

```python
pd.merge(df_1, df_6, left_on='employee', right_on='name').drop('name', axis=1)
```

#### The left_index and right_index keywords

```python
df_1a = df_1.set_index('employee')
df_2a = df_2.set_index('employee')
display('df_1a', 'df_2a')
```

```python
pd.merge(df_1a, df_2a, left_index=True, right_index=True)
```

#### JOIN Method

```python
df_1a.join(df_2a)
```
