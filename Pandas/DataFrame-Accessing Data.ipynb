### access/select dataframe by row & columns
>Access a group of rows and columns by label(s) or a boolean array.

>format .LOC [ Rows, Columns ]

```python
ufo.loc[0] # select only row zero
```

```python
ufo.loc[0,:] # select row Zero and all columns
```

```python
ufo.loc[0:3,:] # select first 4 rows and all columns
```

```python
#selecting multiple row we use brackets
ufo.loc[[0,2,4],:] # selecting rows 0&2&4 and all columns
```

```python
ufo.loc[[0,2,4],['City', 'State']] # selecting multiple rows and multiple columns
```

```python
ufo.loc[:5,'City' : 'Shape Reported']
```

```python
ufo.loc[20:25,:]
```

#####  pandas.DataFrame.iloc

```python
ufo.iloc[[1,5,8],:]
```

```python
ufo.iloc[[2,4,6],[1,3]]
```
