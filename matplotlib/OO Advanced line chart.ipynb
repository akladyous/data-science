```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(0, 2*np.pi, 5000)
```

```python
fig = plt.figure(figsize=[10,10])
```

```python
fig.tight_layout(pad=0.1, w_pad=0.5, h_pad=1.0)
fig.legend(loc='upper left', labels=['axis1', 'axis2'])
fig.subplots_adjust(wspace=0.25, hspace=0.25)
fig.suptitle('this is the figure title', fontsize=12)
```

```python
ax1=fig.add_subplot(2,2,1)
ax1.set_xlabel('X axis 1')
ax1.set_ylabel('Y axis 1')
ax1.set_title('plot 1')

# ax1.set(xlim=[y1.min()+1, y1.max()+1])
#ax1.axis([x1.min(), x1.max(), y1.min()+1, y1.max()+1])

ax1.plot(x,np.sin(x*2), color='r', label='plot 0')
ax1.legend(loc='upper left')
```

```python
ax2=fig.add_subplot(2,2,2)
ax2.set_xlabel('X axis 2')
ax2.set_ylabel('Y axis 2')
ax2.set_title('plot 2')

ax2.plot(x,np.cos(x*3), color='g', label='plot 1')
ax2.legend(loc='upper left')
```

##### 

```python
ax3=fig.add_subplot(2,2,3)
ax3.set_xlabel('X axis 2')
ax3.set_ylabel('Y axis 2')
ax3.set_title('plot 3')

ax3.plot(x,np.sin(x*4), color='b', label='plot 3')
ax3.legend(loc='upper left')
```

```python
ax4=fig.add_subplot(2,2,4)
ax4.set_xlabel('X axis 2')
ax4.set_ylabel('Y axis 2')
ax4.set_title('plot 4')

ax4.plot(x,np.cos(x*4), color='k', label='plot 4')
ax4.legend(loc='upper left')
```
