```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate categorical data (0 for male, 1 for female)
np.random.seed(0)  # for reproducibility
gender = np.random.choice([0, 1], size=100)
```

```python
probability_male = np.sum(gender == 0) / np.size(gender)
probability_female = np.sum(gender == 1) / np.size(gender)
print("Probability of Male:", probability_male)
print("Probability of Female:", probability_female)
```

```python
# Calculate the log-odds
log_odds_male = np.log(probability_male / (1 - probability_male))
log_odds_female = np.log(probability_female / (1 - probability_female))

print("Log-Odds of Male:",   log_odds_male)
print("Log-Odds of Female:", log_odds_female)
```

```python
# Plot the log-odds
labels = ["Male", "Female"]
log_odds = [log_odds_male, log_odds_female]

plt.bar(labels, log_odds)
plt.xlabel("Gender")
plt.ylabel("Log-Odds")
plt.title("Log-Odds of Gender")
plt.show()
```

```python
x = np.linspace(0.001, 0.999, 100)  # Avoiding exactly 0 and 1 to prevent division by zero

# Calculate the logit function
logit = np.log(x / (1 - x))

# Plot the logit function
plt.plot(x, logit)
plt.xlabel("Probability (x)")
plt.ylabel("Logit (log-odds)")
plt.title("Plot of Logit function (log-odds) in the domain of 0 to 1")
plt.grid(True)
plt.show()
```
