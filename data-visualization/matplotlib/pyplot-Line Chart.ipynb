```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
def make_xy(xy_size):
    x=[]
    y=[]
    for i in range(int(xy_size)):
        x.append(i)
        y.append(np.random.randint(int(xy_size)))
    return x,y
```

#### [Line Style - link to matplotlib.org-> linestyle](https://matplotlib.org/3.3.3/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D.set_linestyle)

```python
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('title x y')

x1, y1 = make_xy(10)
plt.plot(x1, y1)

x2, y2 = make_xy(20)
plt.plot(x2, y2)
plt.legend(['X legend', 'Y legend'])
plt.legend(['plot one', 'plot two'])
plt.show()
```

#### PLOT passing arguments: 'marker', 'label', 'linestyle', 'color'

```python
x1, y1 = make_xy(20)
line1 = plt.plot(x1, y1, 'b--', marker='+',label='PLT1 legend')

x2, y2 = make_xy(20)
plt.plot(x2, y2, marker='o', label='PLT2 legend', linestyle='-', color='g')
#plt.legend(['X legend', 'Y legend'])
plt.legend()
plt.axis([0, 25, 0, 25])
plt.grid(True)
plt.show()
```

```python
plt.style.use('default')

x = np.linspace(1,20,100)
y1 = np.cos(x)
y2 = np.sin(x)
plt.plot(x, y1, 'b-',label='PLT1 legend')

plt.plot(x, y2, label='PLT2 legend', linestyle='--', color='r')
#plt.legend(['X legend', 'Y legend'])
plt.legend()
plt.grid(True)
plt.show()
```
