# Standard Normal Distribution

The standard normal distribution, also known as the Z-distribution, is a probability distribution that is often used in statistical analysis. It is a continuous distribution with a mean of 0 and a standard deviation of 1. The Z-distribution is useful in hypothesis testing, confidence interval estimation, and determining probabilities associated with specific values or ranges of values.

1. Z-Score formula with known population parameter:
    
    The Z-score is indeed used to assess the relative position of a single data point (value) within a distribution, typically a sample.
    
    The formula for calculating the Z-score when the population parameter is known is
    
     $[ Z = \frac{{X} - \mu}{\sigma}]$
    
    Where:
    
    - Z is the Z-score
    - X is the value we want to calculate the Z-score for
    - Œº is the population mean
    - œÉ is the population standard deviation
2. Z-Statistics formula with unknown population parameter:
    
    z-statistics are used to test hypotheses about population means when the population standard deviation is known.
    
    They are calculated as the difference between the sample mean and the population mean, divided by the standard deviation of the sampling distribution. Z-statistics are based on the normal distribution and are used to calculate p-values for hypothesis tests.
    
    When the population parameter is unknown, we use the sample mean and sample standard deviation as estimates. The formula for calculating the Z-score in this case is:
    
    $[ Z = \frac{\bar{X} - \mu}{\frac{\sigma}{\sqrt{n}}}]$
    
    Where:
    
    - Z is the Z-score
    - X is the value we want to calculate the Z-score for

    - s is the sample standard deviation
    - n is the sample size

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
mu = 0
sigma = 1

x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 1000)
pdf = norm.pdf(x, mu, sigma)

plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label="Normal Distribution")
plt.title("Z-Distribution (Standard Normal Distribution)")
plt.xlabel("Z-Score")
plt.ylabel("Probability Density Function (PDF)")
plt.legend()
plt.grid(True)
plt.show()
```

**Z-Score:**
- **Definition:** The z-score, also known as the standard score, is a statistical measure that describes a value's relationship to the mean of a group of values, measured in terms of standard deviations from the mean.

- **Formula:** The formula for calculating the z-score of a data point $(x)$ in a distribution with mean $(\mu)$ and standard deviation $(\sigma)$ is given by:  
$[ z = \frac{(x - \mu)}{\sigma} ]$

- **Interpretation:** A positive z-score indicates a data point above the mean, while a negative z-score indicates a data point below the mean. The magnitude of the z-score tells you how far away the data point is from the mean in terms of standard deviations.

**Standardization**
- Standardization is a process that transforms a dataset into a standard normal distribution, often referred to as the Z-distribution. This transformation allows for meaningful comparisons between different datasets. In a standard normal distribution, the mean $(\mu)$ is 0, and the standard deviation $(\sigma)$ is 1.



```python
mu = 5
sigma = 2
rv_normal_distribution = norm.rvs(loc=mu, scale=sigma, size=1000)

rv_mean = np.mean(rv_normal_distribution)
rv_sigma = np.std(rv_normal_distribution)


min_rv, max_rv = rv_normal_distribution.min(), rv_normal_distribution.max()
x_normal = np.linspace(min_rv, max_rv, 1000)
pdf_normal = norm.pdf(x_normal, rv_mean, rv_sigma)  # Corrected line

# Normalize the data
z_distribution = [(x - rv_mean) / rv_sigma for x in rv_normal_distribution]
x_axis_range_standardized = np.linspace(-3, 3, 1000)
pdf_standardized = norm.pdf(x_axis_range_standardized, 0, 1)  # Standard normal distribution

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(rv_normal_distribution, bins=30, color='blue', alpha=0.7, density=True)
plt.plot(x_normal, pdf_normal, "r--", linewidth=2, label="PDF (normal)")  # Corrected line
plt.title('Normal Distribution')
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.legend()

plt.subplot(1, 2, 2)
plt.hist(z_distribution, bins=30, color='green', alpha=0.7, density=True)
plt.plot(x_axis_range_standardized, pdf_standardized, "r--", linewidth=2, label="PDF (standardized)")
plt.title('Standardized Distribution')
plt.xlabel('Standardized Values')
plt.ylabel('Probability Density')
plt.legend()

plt.tight_layout()
plt.show()
```

## Z-Statistic:

- **Definition:** The z-statistic is a measure used in hypothesis testing when the population standard deviation is known. It is used to determine whether there is a significant difference between the sample mean and the population mean.

- **Formula:** The formula for calculating the z-statistic for a sample mean $(\bar{x})$ in a population with known standard deviation $(\sigma)$ and sample size $(n)$ is given by: $[z = \frac{(\bar{x} - \mu)}{(\sigma / \sqrt{n})}]$

- **Interpretation:** The z-statistic is compared to critical values from the standard normal distribution to make inferences about the population mean. If the calculated z-statistic falls in the critical region, it suggests that there is a significant difference between the sample mean and the population mean.

```python
# Set parameters for the population distribution
population_mean = 5
population_std = 2

# Generate a random sample from the normal distribution
sample_size = 100
random_sample = norm.rvs(loc=population_mean, scale=population_std, size=sample_size)

# Calculate sample mean and standard deviation
sample_mean = np.mean(random_sample)
sample_std = np.std(random_sample, ddof=1)  # ddof=1 for sample standard deviation

# Calculate Z-statistics
z_statistics = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))

# Display the results
print(f"Population Mean: {population_mean}")
print(f"Sample Mean: {sample_mean}")
print(f"Sample Standard Deviation: {sample_std}")
print(f"Z-Statistics: {z_statistics}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mu = 0
sigma = 1

# Generate data for the standard normal distribution
x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 1000)
pdf = norm.pdf(x, mu, sigma)

# Find the z-score corresponding to a cumulative probability of 0.05
z_score_lower_bound = norm.ppf(0.05)
z_score_upper_bound = norm.ppf(0.95)

# Plot the standard normal distribution with shaded area
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label="Normal Distribution")
plt.fill_between(x, pdf, where=(x <= z_score_lower_bound), color='skyblue', alpha=0.4, label="Shaded Area")
plt.fill_between(x, pdf, where=(x >= z_score_upper_bound), color='skyblue', alpha=0.4)

# Highlight the z-score point
plt.scatter([z_score_lower_bound], [norm.pdf(z_score_lower_bound, mu, sigma)], color='red', label=f'Z-Score: {z_score_lower_bound}', zorder=5)
plt.scatter([z_score_upper_bound], [norm.pdf(z_score_upper_bound, mu, sigma)], color='red', label=f'Z-Score: {z_score_upper_bound}', zorder=5)

plt.title("Z-Distribution (Standard Normal Distribution)")
plt.xlabel("Z-Score")
plt.ylabel("Probability Density Function (PDF)")
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```
