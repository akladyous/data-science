```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# Parameters for the t-distribution
mu = 30
sigma = 5
sample_size = 1000
degrees_of_freedom = sample_size - 1

# Generate a sample of ages
ages = np.random.normal(mu, sigma, sample_size)

# Calculate the t-distribution values
x_values = np.linspace(min(ages), max(ages), sample_size)
pdf_values = stats.t.pdf(x_values, degrees_of_freedom, mu, sigma)

# Plot the histogram of the dataset
plt.hist(ages, bins=30, density=True, alpha=0.7, color='blue', label='Age Distribution')

# Plot the t-distribution curve
plt.plot(x_values, pdf_values, 'r--', label=f't-Distribution (df={degrees_of_freedom})')

plt.title('t-Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Probability Density')
plt.legend()

plt.show()
```
