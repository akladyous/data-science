# Normal Distribution

The normal distribution, also known as the Gaussian distribution, is a continuous probability distribution that is symmetric and bell-shaped. 
It is characterized by its mean and standard deviation, which determine the center and spread of the distribution, respectively. The shape of the normal distribution is often referred to as the "bell curve" due to its characteristic shape.

# The Probability Density Function (PDF)

The Probability Density Function (PDF) is a statistical function that describes the likelihood of a continuous random variable taking on a specific value. In the context of a continuous probability distribution, the PDF provides the probability density at each possible outcome.

For a continuous random variable $(X)$, the PDF is denoted as $(f(x))$, and it satisfies the following properties:

**1. Non-Negativity:** $(f(x) \geq 0)$ for all $(x)$ in the range of possible values.

**2. Area under the Curve:** The total area under the PDF curve over the entire range of possible values is equal to 1.

Mathematically, the PDF is used to model the distribution of the random variable. For a normal distribution with mean \(\mu\) and standard deviation \(\sigma\), the PDF is given by:

$ [ f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{1}{2}\left(\frac{x - \mu}{\sigma}\right)^2} ] $

This formula provides the probability density at each point \(x\) in the distribution. It doesn't give the probability of a specific value occurring (as the variable is continuous), but rather the relative likelihood of values in different ranges.

In summary, the Probability Density Function is a fundamental concept in probability theory and statistics, particularly for continuous random variables. It helps us understand the distribution of probabilities across the possible outcomes of a continuous variable.

****Empirical Rule****

The empirical rule, also known as the 68-95-99.7 rule, is a rule of thumb that describes the approximate proportion of data that falls within certain standard deviation intervals of a normal distribution. According to this rule:

- Approximately 68% of the data falls within one standard deviation of the mean.
- Approximately 95% of the data falls within two standard deviations of the mean.
- Approximately 99.7% of the data falls within three standard deviations of the mean.

This rule provides a useful guideline for understanding the spread of data in a normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# x = np.linspace(-4, 4, 1000)
mu, sigma = 0, 1                    # mean and standard deviation
x = np.linspace(
    norm(mu, sigma).ppf(0.01) - 1,  # z-score - Start plotting here
    norm(mu, sigma).ppf(0.99) + 1,  # z-score - End plotting here
    1000,  # Number of points
)
y = norm(mu, sigma).pdf(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label="PDF of Normal Distribution")
plt.title("Probability Density Function (PDF) of Normal Distribution")
plt.xlabel("Random Variable (x)")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

The probability density function (PDF) of a normal distribution is given by:

$ [ f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{(x - \mu)^2}{2\sigma^2}} ] $

Where:
- $(f(x))$ is the PDF representing the probability density at a specific value $(x)$.
- $(\mu)$ is the mean of the normal distribution.
- $(\sigma)$ is the standard deviation of the normal distribution.
- $(\pi)$ This is the mathematical constant representing the ratio of a circle's circumference to its diameter, approximately 3.14159.
- $(e)$ is the mathematical constant approximately equal to 2.71828.


```python
mu, sigma = 0, 2
rv_normal_distribution = norm.rvs(loc=mu, scale=sigma, size=1000)

bins_square_root_rule = int(np.sqrt(len(rv_normal_distribution)))
plt.figure(figsize=(8, 6))
plt.hist(
    rv_normal_distribution, bins=bins_square_root_rule, density=True, alpha=0.6, color="g", label="Generated Samples"
)

# Plot the probability density function (PDF) of the normal distribution

# Using np.linspace(xmin, xmax, 100) allows you to create a smooth range of values along the x-axis for plotting the probability density function (PDF) curve.
# The idea is to have a range that spans a bit beyond the minimum and maximum values in your dataset to ensure that the entire PDF curve is plotted.
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
pdf = norm.pdf(x, mu, sigma)
plt.plot(x, pdf, "b-.", linewidth=2, label="PDF (plt.xlim())")


# otherwise we can simply use min and max(samples) to get the minimum and maximum values in your dataset.
xmin, xmax = min(rv_normal_distribution), max(rv_normal_distribution)
x = np.linspace(xmin, xmax, 100)
pdf = norm.pdf(x, mu, sigma)
plt.plot(x, pdf, "r--", linewidth=2, label="PDF (min/max)")


plt.title("Histogram of Random Samples from Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Probability Density")

plt.legend()
plt.grid(True)
plt.show()
```
