## Probability Mass Function (PMF)

The Probability Mass Function (PMF) is a concept in probability theory and statistics that describes the probability distribution of a **discrete random variable**.

It assigns probabilities to the possible outcomes of the random variable, indicating the likelihood of each outcome occurring.

For a discrete random variable $(X)$, the PMF is denoted as $(P(X = x))$, representing the probability that the random variable takes on a specific value $(x)$. The PMF must satisfy two conditions:

1. **Non-negativity:** $(P(X = x) \geq 0)$  for all values of $(x)$.
2. **Summation:** The sum of probabilities over all possible values of $(X)$ must equal 1: $\sum P(X = x) = 1)$.

Mathematically, the PMF is defined as follows:

$[P(X = x) = f(x)]$

where \(f(x)\) is the function that maps each value $(x)$ to its probability.

Here are a few key points about the PMF:

- The PMF is specific to discrete random variables, which take on distinct, separate values.
- It provides a complete description of the probability distribution for the random variable.
- The probabilities assigned by the PMF indicate the likelihood of observing each possible outcome.


### Example: Fair Six-Sided Die

Suppose we have a fair six-sided die. The PMF for the outcome of rolling the die is:

$[ P(X = 1) = P(X = 2) = P(X = 3) = P(X = 4) = P(X = 5) = P(X = 6) = \frac{1}{6} ]$

This satisfies the conditions of the PMF, as each probability is non-negative, and the sum of all probabilities is 1.

In practice, the PMF is often presented as a table or a formula that lists the probabilities for each possible outcome of the random variable.


- Event Definition:
  - The event was defined as the outcome of rolling a fair six-sided die.
- Number of Ways It Can Happen:
  - Since the die is fair, each outcome (1, 2, 3, 4, 5, 6) is equally likely.
  - The number of ways each outcome can happen is 1, as there is only one way for each face to appear.
- Total Number of Outcomes:
  - The total number of outcomes corresponds to the number of faces on the die, which is 6.
- Application of Probability Formula:
  - Using the probability formula:
    $ [ P(\text{Event}) = \frac{\text{Number of ways it can happen}}{\text{Total number of outcomes}} ] $
    , we calculated the probability for each outcome.
- Probability Mass Function (PMF):
  - The Probability Mass Function (PMF) was used to represent the probability distribution of the discrete random variable (the outcome of rolling the die).
  - The PMF values were set uniformly, each being $​\frac{1}{6}$ since the die is fair.
- Visualization:
  - We created a stem plot using Matplotlib to visualize the PMF.
  - Each stem in the plot represented one of the six possible outcomes, and the height of the stem indicated the probability of that outcome.
- Result:
  - The resulting plot accurately depicted the fair die experiment, showing that each outcome has an equal probability of $​\frac{1}{6}$, consistent with the fairness of the die.

In summary, the fair die experiment demonstrated the calculation and visualization of probabilities for each possible outcome using the probability formula, and the resulting PMF accurately represented the fairness of the die.

### Bernoulli distribution

The Bernoulli distribution is specifically designed for scenarios with two possible outcomes, typically labeled as success (1) and failure (0).

The probabilities of success and failure are denoted by parameters $p$ and $1-p$ respectively.

```python
prob_success = 0.3  # Probability of success
samples = np.random.choice([0, 1], size=1000, p=[1-prob_success, prob_success])

probabilities = stats.bernoulli.pmf(k=samples, p=prob_success)
probability_success = stats.bernoulli.pmf(k=1, p=prob_success)
probability_failure = stats.bernoulli.pmf(k=0, p=prob_success)

print(probability_success)
print(probability_failure)

plt.hist(probabilities, bins=[-0.5, 0.5, 1.5], edgecolor='black', align='mid', rwidth=0.8)
plt.title('Bernoulli Distribution')
plt.xlabel('Outcome')
plt.ylabel('Frequency')
plt.xticks([0, 1], ['Failure (0)', 'Success (1)'])
plt.show()
```

```python
possible_outcomes = np.arange(1, 7)

# Probability Mass Function (PMF) for a fair six-sided die
pmf_values = np.ones_like(possible_outcomes) / 6

plt.stem(possible_outcomes, pmf_values, basefmt="b", markerfmt="bo")
plt.title("Probability Mass Function (PMF) of a Fair Six-Sided Die")
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.xticks(possible_outcomes)
plt.show()
```

### Probability Mass Function (PMF) for a fair coin flip

```python
# Define the possible outcomes of the coin flip
possible_outcomes = ["Heads", "Tails"]

# Probability Mass Function (PMF) for a fair coin flip
pmf_values = np.ones(len(possible_outcomes)) / len(possible_outcomes)
print(pmf_values)

# Plotting the PMF
plt.stem(possible_outcomes, pmf_values, basefmt="b", markerfmt="bo")
plt.title("Probability Mass Function (PMF) of a Fair Coin Flip")
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.show()
```

```python
outcomes = ["head", "tail"]
num_flips = 100

sample_space = np.random.choice(outcomes, size=num_flips)
unique_flips, counts_flips = np.unique(sample_space, return_counts=True)

# Define the probability of getting heads (success)
p_heads = counts_flips[outcomes.index("head")] / num_flips
p_tails = counts_flips[outcomes.index("tail")] / num_flips

probability_heads = stats.bernoulli.pmf(k=1, p=p_heads)  # Probability of 'success' (1)
probability_tails = stats.bernoulli.pmf(k=0, p=p_heads)  # Probability of 'failure' (0)

print(f"Probability of getting heads: {probability_heads:.3f}")
print(f"Probability of getting tails: {probability_tails:.3f}")

# Plotting the results
plt.bar(unique_flips, counts_flips, color=["blue", "green"])
plt.title(f"Coin Flip Experiment (Bernoulli Distribution)")
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.xticks(outcomes)

plt.text(
    0,
    counts_flips[0] + 1,
    f"{probability_tails:.2%}",
    ha="center",
    va="center",
    color="black",
)
plt.text(
    1,
    counts_flips[1] + 1,
    f"{probability_heads:.2%}",
    ha="center",
    va="bottom",
    color="black",
)
plt.ylim(top=max(counts_flips) + 5)
plt.show()
```

```python
prob_of_success = 0.8
rv = stats.bernoulli(prob_of_success)
unfair_game = {"win": rv.pmf(k=1), "lose": rv.pmf(k=0)}

rounded_unfair_game = {key: round(value, 3) for key, value in unfair_game.items()}

fig, ax = plt.subplots()
bars = ax.bar(rounded_unfair_game.keys(), rounded_unfair_game.values(), color=['green', 'orange'])

for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, yval, f"{yval * 100:.1f}%", ha='center', va='bottom')

ax.set_title("Unfair Game Probabilities")
ax.set_ylabel("Probability")
ax.set_ylim(0, 1)
ax.set_xlabel("Outcome")
plt.show()
```

```python
prob_of_success = 0.7
rv = stats.bernoulli(prob_of_success)
unfair_game = {"win": rv.pmf(k=1), "lose": rv.pmf(k=0)}

rounded_unfair_game = {key: round(value, 3) for key, value in unfair_game.items()}
percentages = {key: value * 100 for key, value in rounded_unfair_game.items()}


fig, ax = plt.subplots()
bars = ax.bar(rounded_unfair_game.keys(), rounded_unfair_game.values(), color=['green', 'orange'], label='Probability')
for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, yval , f"{yval * 100:.1f}%", ha='center', va='bottom')


cumulative_probs = [rv.cdf(0), rv.cdf(1)]
ax.bar(rounded_unfair_game.keys(), height=cumulative_probs, color='blue', alpha=0.5, label='Cumulative Probability')

ax.set_title("Unfair Game Probabilities")
ax.set_ylabel("Probability")
ax.set_ylim(0, 1.02)
ax.set_xlabel("Outcome")
ax.legend()
plt.show()
```

```python
low = 1
high = 6
rv = stats.randint(low, high + 1)
sample = rv.rvs(size=1000)

values = np.arange(low, high + 1)
pmf_values = rv.pmf(values)

fig, ax = plt.subplots()
ax.vlines( values, 0, pmf_values, colors='b', lw=15, alpha=0.5, label='PMF')
ax.hist(sample, bins=range(low, high + 2), density=True, rwidth=0.8, color='lightgray', edgecolor='black', alpha=0.7, label='Sample', align='mid')

ax.set_title('Uniform Discrete Distribution')
ax.set_xlabel('Values')
ax.set_ylabel('Probability')
ax.legend()
ax.set_ylim(0, 0.25)
ax.set_xticks([val + 0.5 for val in values])
ax.set_xticklabels(values)

plt.show()
```

```python
from scipy.stats import randint

low = 1
high = 6

pmf_rolls = {}
cdf_rolls = {}

for roll in range(low, high + 1):
    pmf_rolls[roll] = randint.pmf(k=roll, low=low, high=high + 1)
    cdf_rolls[roll] = randint.cdf(roll, low=low, high=high + 1)

print("PMF Rolls:", pmf_rolls)
print("CDF Rolls:", cdf_rolls)
```

```python
low = 1
high = 6

pmf_rolls = {}
cdf_rolls = {}

for roll in range(low, high + 1):
    pmf_rolls[roll] = randint.pmf(k=roll, low=low, high=high + 1)
    cdf_rolls[roll] = randint.cdf(roll, low=low, high=high + 1)

plt.figure(figsize=(6, 6))

plt.bar(pmf_rolls.keys(), pmf_rolls.values(), 0.8, color='b', alpha=0.5, label='Uniform Discrete Probability')
plt.bar(cdf_rolls.keys(), cdf_rolls.values(), 0.8, bottom=list(pmf_rolls.values()), color='g', alpha=0.5, label='Cumulative Probability')

plt.ylabel('Probability')
plt.xlabel('Result of Die Roll')
plt.legend()
plt.show()
```
