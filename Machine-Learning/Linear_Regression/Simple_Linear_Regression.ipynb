```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn import datasets
from scipy import stats
%matplotlib inline
```

```python
np.random.seed(5)
X = np.arange(20, 60) # AGEs 
Y = np.sort(np.random.randint(25000, 35000, size=40))# INCOMEs
```

```python
X_mean = np.mean(X)
X_std = np.std(X)
Y_mean = np.mean(Y)
Y_std = np.std(Y)
n=X.size
```

#Least-Squares Regression

The most common method for fitting a regression line is the method of least-squares. This method calculates the best-fitting line for the observed data by minimizing the sum of the squares of the vertical deviations from each data point to the line (if a point lies on the fitted line exactly, then its vertical deviation is 0). Because the deviations are first squared, then summed, there are no cancellations between positive and negative values.

<!-- ![slope](../img/slope_intercept.gif) -->

```python
nominator   = 0
denominator = 0

for i in range(X.size):
    nominator   += (X[i] - X_mean) * (Y[i] - Y_mean)
    denominator += (X[i] - X_mean)**2
#calculate the slope
M = nominator / denominator
```

```python
#calculate the intercept
```

```python
C = Y_mean - M * X_mean
```

```python
C
```

```python
Y_predected = M * X + C
```

## Pearsons correlation
>now we can determine the correlation value between the 2 variables and the way we do it is by using the Pearson correlation coefficient equation 'r'. this value can have a range of values from +1 to -1.

>The stronger the association of the two variables, the closer the correlation coefficient 'r' will be to +1 or -1 depending on whether the relationship is positive or negative, respectively.
A value of 0 indicates that there is no association between the two variables.

```python
corr, _ = stats.pearsonr(X, Y)
```

```python
fig = plt.figure(figsize=[15,5])
ax1=fig.add_subplot()
ax1.set_title('Simple Linear Regression')
ax1.set_xlabel("Age")
ax1.set_ylabel('Income')
ax1.scatter(X, Y)
ax1.text(X.min(),Y.max()-1000,'Pearsons correlation: %.3f'% corr, fontsize=15, c='g')
ax1.plot(X, Y_predected, c='r')
```
