```python
a1=np.arange(1,4).reshape((1,3)) # AXIS 0
a1
```

```python
a2=np.arange(1,4).reshape(3,1)   # AXIS 1
a2
```

```python
a3=np.arange(1,10).reshape((3,3))
print(a3)
print(array_info(a3))
```

```python
a4=np.arange(9,0,-1).reshape((3,3))
print(a4)
print(array_info(a4))
```

```python
a1+a2 # addition by AXIS 0
```

```python
a2+a3 # addition by AXIS 1
```

```python
a3+a4
```
